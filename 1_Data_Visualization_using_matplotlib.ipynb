```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/dbda.STUDENTSDC/Documents/Advanced Analytics using Statistics/tips.csv")
```

```python
# 1. Histogram of total_bill - Shows the distribution of a single variable(here, total bill)
plt.figure(figsize =(6,4))
plt.hist(df['total_bill'], bins = 10, color ='skyblue',edgecolor = 'black')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

# Same plot using seaborn
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], bins = 10 , kde = True, color = 'darkgreen') # kde = Kernel Density Estimation- Shows a continous line over the histogram 
plt.title('Distribution of total bill')
plt.show()
```

```python
# 2. Bar plot of average tip by gender
avg_tip_gender = df.groupby('gender')['tip'].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_tip_gender.index, avg_tip_gender.values, color = ['pink','lightblue'])
plt.title('Average Tip by Gender')
plt.ylabel('Average Tip')
plt.show()
```

```python
# Using Seaborn
avg_tip_gender = df.groupby('gender')['tip'].mean().reset_index()
plt.figure(figsize = (6,4))
sns.barplot(data=avg_tip_gender, x='gender',y='tip',palette=['pink','lightblue'])
plt.title('Average Tip by Gender')
plt.ylabel('Average Tip')
plt.show()
```

```python
# 3. Scatter plot of total_bill vs tip- Tells us how these are related to each other
plt.figure(figsize=(6,4))
plt.scatter(df['total_bill'],df['tip'], color ='orange' ,edgecolor = 'black')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# With Seaborn, but also with regression line
plt.figure(figsize=(6,4))
sns.regplot(x ='total_bill', y ='tip',data =df, color ='red')
plt.title('Total Bill vs Tip With Regression Line')
plt.show()
```

```python
# 4. Box plot of tip by gender using sns-Useful for identifying "outliers"
plt.figure(figsize=(6,4))
sns.boxplot(x = 'gender', y='tip',data = df, hue ='gender', palette = 'pastel', legend =False)
plt.title('Tip by Gender')
plt.show()
```

```python
# 5. Count plot of smokers vs non-smokers
plt.figure(figsize=(6,4))
sns.countplot(x ='smoker',data=df, hue = 'smoker', palette = 'Set2', legend= False)
plt.title('Count of Smokers vs non-Smokers')
plt.show()
```
