```python
# Creating multi-dimensional arrays as "Structure"
import numpy as np
```

```python
students = np. array([('Alice',95,20),
                      ('Bob',88,21),
                      ('Charlie',92,19)],dtype = [('name','U10'),('marks','i4'),('age','i4')])
```

```python
print(students) 
```

```python
print(students['name'])
```

```python
print(students['marks'])
```

```python
print(students['age'])
```

```python
print(students[students['name'] == 'Alice']['marks'])
```

```python
print(students[students['name'] == 'Alice']['marks'][0])
```
