```python
#Indexing, Slicing, Filtering numpy arrays
import numpy as np
```

```python
marks = np.array([55,68,72,49,83,91,37,65,78,59,88])
print(marks)
```

```python
# indexing and slicing
print(marks[8]) # 9th student
print(marks[1:5])
print(marks[0:5])
```

```python
# Broadcasting (Update multiple values)
marks[0:3] = 75
print(marks)
```

```python
#Vectorization
result = marks + 5
print(result)
```

```python
# Reset Array
marks = np.array([55,68,72,49,83,91,37,65,78,59,88])
```

```python
# Slices and Copy
top_batch = marks[0:5] # "View" or "Shallow copy"
print(top_batch)
```

```python
# Modify slice (affects original)
top_batch[:] = 99
print(top_batch)
print(marks)
```

```python
# To avoid this,use copy()
marks = np.array([55,68,72,49,83,91,37,65,78,59,88])
marks_copy = marks.copy() # deep copy
marks_copy[3] = 99
print(marks_copy)
print(marks)
```

```python
# Conditional Selection
marks = np.array([69,26,78,91,54,34])
print(marks)
```

```python
passed = marks >= 35 # boolean array
print(passed)
print(marks[passed]) #actual value
```
