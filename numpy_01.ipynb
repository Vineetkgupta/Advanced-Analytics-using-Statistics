```python
#Special Arrays
zeros = np.zeros((2,3))
ones = np.ones((3,2))
identity = np.eye(3)
print(zeros)
print(ones)
print(identity)
```

```python
empty = np.empty((2,2)) # uninitialized garbage float array
full = np.full((2,2),7)
print(empty)
print(full)
```

```python
# Creating arrays with range
arr = np.arange(55,65) # start: 55, end:64
print(arr, arr[0] , arr[5], arr[9])
arr[4] = 99
print(arr[4])
```

```python
run_rates = np.arange(4,11,1)
print(run_rates)
```

```python
#linspace
batting_avg = np.linspace(30,50,5) # here 5 is not a step it is the 5 evenly spread number (total 5)
print(batting_avg)
```

```python
strike_rates = np.random.uniform(100,200,10)
print(strike_rates)
```

```python
runs = np.random.randint(1,151,(2,3))
print(runs)
```

```python
batting_avg = np.random.normal(40,5,(2,3))
print(batting_avg)
```

```python
# Adding new values to an existing array 
arr = np.arange(0,10)
new_arr_single = np.append(arr,10)
new_arr_multiple = np.append(arr, [10,11,12])
print(arr)
print(new_arr_single)
print(new_arr_multiple)
```

```python
#Reshaping
marks = np.arange(50, 62)
print(marks)
reshaped = marks.reshape(3,4)
print(reshaped)
```
